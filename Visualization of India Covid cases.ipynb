```python
#renames the column active to active cases
```

```python
train_df = train.rename(columns = {'Active':'Active Cases'})
```

```python
train_df
```

```python
#import libray for data visualization which is seaborn
```

```python
import seaborn as sns
```

```python
#get a line plot for your active cases and deaths
```

```python
sns.set(rc = {'figure.figsize':(10,9)})
sns.lineplot(x = train_df['Active Cases'], y = train_df['Deaths'])
```

```python
#gets a lineplot for your total  cases and Deaths
```

```python
sns.set(rc = {'figure.figsize':(10,9)})
sns.lineplot(x = train_df['Total Cases'], y = train_df['Deaths'])
```

```python
#renames the state/uts to states
```

```python
trained_df = train_df.rename(columns = {'State/UTs':'States'})
```

```python
trained_df
```

```python
# plots a bar chart for active ratio and states
```

```python
sns.set(rc = {'figure.figsize':(20,20)})
sns.barplot(x = trained_df['Active Ratio (%)'], y = trained_df['States'])
```

```python
#plots a barchat for death ratio and states
```

```python
sns.set(rc = {'figure.figsize':(20,20)})
sns.barplot(x = trained_df['Death Ratio (%)'], y = trained_df['States'])
```
